```python
import re
def remove_pattern(input_txt, pattern):
    r = re.findall(pattern, input_txt)
    for i in r:
        input_txt = re.sub(i, "", input_txt)
    return input_txt
```

```python
df['comment'] = np.vectorize(remove_pattern)(df['Instagram Comment Text'],"<[\w]*")
df.head()
```

```python
data_clean = data_clean.astype({'Sentiment' : 'category'})
data_clean = data_clean.astype({'comment' : 'string'})
data_clean.dtypes
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfVectorizer

tf = TfidfVectorizer()
text_tf = tf.fit_transform(data_clean['comment'].astype('U'))
text_tf
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(text_tf,data_clean['Sentiment'],test_size=0.2,random_state=42)
```

```python
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix

clf = MultinomialNB().fit(x_train,y_train)
predicted = clf.predict(x_test)
print("MultinomialNB Accuracy :", accuracy_score(y_test, predicted))
print("MultinomialNB Precission :", precision_score(y_test, predicted, average="binary", pos_label="negative"))
print("MultinomialNB Recall :", recall_score(y_test, predicted, average="binary", pos_label="negative"))
print("MultinomialNB f1_score :", f1_score(y_test, predicted, average="binary", pos_label="negative"))

print(f'confsion matrix:\n {confusion_matrix(y_test, predicted)}')
print('==================================================')
print(classification_report(y_test, predicted,zero_division=0))
```
